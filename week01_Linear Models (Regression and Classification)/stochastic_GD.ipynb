# 📈 Linear Regression with Stochastic Gradient Descent (SGD)

In this notebook we implement **Linear Regression using Stochastic Gradient Descent (SGD)**.

Goal:  
Predict **exam score** from **hours studied**.

Unlike normal gradient descent, **SGD updates the model using one random data point at a time**.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1️⃣ Load the Dataset

We load the dataset using **pandas**.

```python
data = pd.read_csv('data.csv')

data.head()
```

## 2️⃣ Extract Features and Target

- **x** = hours studied (input)
- **y** = exam score (output)

```python
x = data['hours_studied'].values
y = data['exam_score'].values
```

## 3️⃣ Visualize the Dataset

```python
plt.scatter(x, y)

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.title("Dataset")

plt.show()
```

## 4️⃣ Initialize Model Parameters

We start with:

- slope = 0  
- intercept = 0  

These values will be updated during training.

```python
slope = 0
intercept = 0

learning_rate = 0.01
epochs = 100
```

## 5️⃣ Stochastic Gradient Descent Training

Instead of using the whole dataset, **SGD randomly selects one data point each step** and updates the parameters.

```python
for epoch in range(epochs):

    random_index = random.randint(0, len(x) - 1)

    x_i = x[random_index]
    y_i = y[random_index]

    y_pred = slope * x_i + intercept

    error = y_pred - y_i

    slope_gradient = error * x_i
    intercept_gradient = error

    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient
```

## 6️⃣ Final Model Parameters

```python
print("Slope:", slope)
print("Intercept:", intercept)
```

## 7️⃣ Plot the Regression Line

```python
y_pred = slope * x + intercept

plt.scatter(x, y)

plt.plot(x, y_pred)

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.title("Linear Regression using SGD")

plt.show()
```
