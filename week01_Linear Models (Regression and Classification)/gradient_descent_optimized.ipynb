# Gradient Descent from Scratch (Optimized)

This notebook demonstrates an **optimized** implementation of gradient descent using NumPy vectorization. The key improvements:

- **Vectorized operations** instead of Python loops (much faster)
- **Efficient matrix operations** using NumPy
- **Better performance** especially with larger datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
data = pd.read_csv('data.csv')
x = data['hours_studied'].values
y = data['exam_score'].values
data.head()
```

## Optimized Loss Function (Vectorized)

Instead of looping through each data point, we use NumPy's vectorized operations.

```python
def loss_function(m, b, x, y):
    """
    Vectorized Mean Squared Error calculation
    """
    y_pred = m * x + b
    return np.mean((y_pred - y) ** 2)
```

## Optimized Gradient Descent (Vectorized)

Key optimization: Calculate gradients using vectorized operations instead of loops.

```python
def gradient_descent(m, b, x, y, learning_rate):
    """
    Vectorized gradient descent - much faster than loop-based version
    """
    N = len(x)
    
    # Vectorized prediction
    y_pred = m * x + b
    
    # Vectorized gradient calculation
    error = y_pred - y
    m_gradient = (2/N) * np.sum(error * x)
    b_gradient = (2/N) * np.sum(error)
    
    # Update parameters
    m = m - learning_rate * m_gradient
    b = b - learning_rate * b_gradient
    
    return m, b
```

## Train the Model

```python
# Initialize parameters
m = 0
b = 0
learning_rate = 0.01
epochs = 1000

# Training loop
for i in range(epochs):
    m, b = gradient_descent(m, b, x, y, learning_rate)
    
    if i % 100 == 0:
        loss = loss_function(m, b, x, y)
        print(f'Epoch {i}: Loss = {loss:.4f}, m = {m:.4f}, b = {b:.4f}')

print(f'\nFinal parameters: m = {m:.4f}, b = {b:.4f}')
```

## Visualize Results

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, m * x + b, color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression using Optimized Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```

## Performance Comparison

The optimized version is significantly faster because:

1. **No Python loops** - NumPy operations are implemented in C
2. **Vectorized operations** - Process entire arrays at once
3. **Better memory usage** - More efficient array operations

For larger datasets, the speedup can be 10-100x faster!
